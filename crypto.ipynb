```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
#Createt the empty lists to store the date
crypto_name_list = []
crypto_market_cap_list = []
crypto_price_list = []
crypto_circulateing_supply_list = []
crypto_symbol_list = []
```

```python
#Create an dateframe to help organize teh data
df = pd.DataFrame()
```

```python
def scrape(date = '20220327/'):
    #Get the URL of the website that we want to scrape
    URL = 'https://coinmarketcap.com/historical/' + date

    #Make request to the website
    webpage = requests.get(URL)

    #Parse the text from the website
    soup = BeautifulSoup(webpage.text, 'html.parser')

    #Get the table row element
    tr = soup.find_all('tr', attrs= {'class':'cmc-table-row'})

    # Create a count variable for the number of crpto currencies that we want to scrape
    count = 0

    # Loop through evrety row to gather the data / information 
    for row in tr:
        # if the count is reached 10 then break out of the loop 
        if count == 10:
            break;
        count = count + 1


        #store the name of the crypto into a a variable
        #Find the "td" element to later get the crypto currency name
        name_column = row.find('td', attrs = { 'class':'cmc-table__cell cmc-table__cell--sticky cmc-table__cell--sortable cmc-table__cell--left cmc-table__cell--sort-by__name'})
        crypto_name = name_column.find('a', attrs = {'class':'cmc-table__column-name--name cmc-link'}).text.strip()
        
        #Store the coin market cap of the cryptocurrencyt or coin into a variable
        crypto_market_cap = row.find('td', attrs = {'class':'cmc-table__cell cmc-table__cell--sortable cmc-table__cell--right cmc-table__cell--sort-by__market-cap'}).text.strip()
        
        #Find the crypto price
        crypto_price = row.find('td', attrs ={'class':'cmc-table__cell cmc-table__cell--sortable cmc-table__cell--right cmc-table__cell--sort-by__price'}).text.strip()
        
        #Find and stroe the crypto suppuly and symbol
        crypto_circulateing_supply_and_symbol = row.find('td', attrs ={'class':'cmc-table__cell cmc-table__cell--sortable cmc-table__cell--right cmc-table__cell--sort-by__circulating-supply'}).text.strip()
        
        #Split the datest to circulateing supply and symbol
        crypto_circulateing_supply = crypto_circulateing_supply_and_symbol.split(' ')[0]
        crypto_symbol = crypto_circulateing_supply_and_symbol.split(' ')[1]


        #Append teh data to the lsit 
        crypto_name_list.append(crypto_name)
        crypto_market_cap_list.append(crypto_market_cap)
        crypto_price_list.append(crypto_price)
        crypto_circulateing_supply_list.append(crypto_circulateing_supply)
        crypto_symbol_list.append(crypto_symbol)
```

```python
#Run the scrape function
#You are also change the date
scrape(date = '20220327/')
```

```python
#Store the data into a dataframe to help organize the data
df['Name'] = crypto_name_list
df['Market_cap'] = crypto_market_cap_list
df['Price'] = crypto_price_list
df['Circulating Supply'] = crypto_circulateing_supply_list
df['Symbol'] = crypto_symbol_list
```
